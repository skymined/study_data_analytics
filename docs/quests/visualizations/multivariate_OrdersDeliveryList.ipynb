```python
import pandas as pd
df_shop = pd.read_csv('ShoppingMallDeliveryWithDate.csv')
df_shop[:5]
```

#### 1. 상품구매금액이 80,000원 이상인 주문들을 찾으세요.

```python
df_shop[['결제수단', '배송번호', '상품구매금액']].query('상품구매금액 >= 80000')
```

#### 2. 모바일웹을 통해 주문된 상품 중, 상품구매금액이 평균 이상인 주문들을 찾으세요.

```python
import numpy as np
money_avg = np.mean(df_shop['상품구매금액'])
money_avg
```

```python
condition_2_1 = '주문경로 == "모바일웹"'
condition_2_2 = f'상품구매금액 >= {money_avg}'
conditions_2 = f"{condition_2_1} and {condition_2_2}"
df_shop[['결제수단', '배송번호', '상품구매금액', '주문경로']].query(conditions_2)
```

#### 3. 2019년 1월 7일 이전에 배송이 시작된 주문들을 찾으세요.

```python
df_shop[['배송번호', '배송시작일', '배송완료일']].query('배송시작일 <= "2019-01-07"')
```

#### 4. 배송 시작일과 배송 완료일이 3일 이상 차이 나는 주문들을 찾으세요.

```python
df_shop[['배송완료일', '배송시작일']].info()
```

```python
df_shop['배송시작일_datetime'] = pd.to_datetime(df_shop['배송시작일'].str.split().str[0], format='%Y-%m-%d')
df_shop['배송완료일_datetime'] = pd.to_datetime(df_shop['배송완료일'].str.split().str[0], format='%Y-%m-%d')
```

```python
df_shop['배송기간'] = (df_shop['배송완료일_datetime'] - df_shop['배송시작일_datetime']).dt.days
df_shop['배송기간']
```

```python
df_shop[['결제수단', '배송번호', '배송시작일', '배송시작일_datetime', '배송완료일', '배송완료일_datetime', '배송기간']].query('배송기간  >= 3')
```

#### 5. 무통장입금을 결제수단으로 사용한 주문 중, 상품번호가 1000번 이상인 주문들을 찾으세요.

```python
condition_5_1 = '결제수단 == "무통장입금"'
condition_5_2 = '상품번호 >= 1000'
```

```python
df_shop[['결제수단', '배송번호', '상품번호']].query(f"{condition_5_1} and {condition_5_2}")
```

#### 6. 상품번호가 50번 이하이거나 1000번 이상인 주문들을 찾으세요.

```python
df_shop['상품번호'].info()
```

#### 7. 주문일이 20181231인 주문 중, 수량이 1개 이상인 주문들을 찾으세요.

```python
df_shop[['주문일', '수량']].info()
```

#### 8. PC쇼핑몰을 통해 주문된 상품 중, 배송번호가 'D-20181231'로 시작하는 주문들을 찾으세요.

#### 9. 상품구매금액이 70,000원 이상 100,000원 이하인 주문들 중, 수량이 1개인 주문들을 찾으세요.

#### 10. 모바일웹을 통한 주문 중, 배송번호가 'D-20181230' 이후로 시작되고 상품구매금액이 50,000원 이상인 주문들을 찾으세요.
