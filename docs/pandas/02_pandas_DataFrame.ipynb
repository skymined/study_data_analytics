## pandas

### DataFrame (데이터 프레임) = (Series + Series + ...) + column

- 다(복합) 변수


### DataFrame 생성

```python
# @title
list = [3,4,5,6]
pd.DataFrame(list)
```

```python
# @title
pd.Series(list)
```

```python
# @title
double_list = [
             [1000, '과자','2019-12-31','반품'],
             [2000, '음료', '2020-03-02', '정상'],
             [3000, '아이스크림', '2020-02-03','정상'],
             [1000,'과자','2019-12-31','반품']
            ]
pd.DataFrame(double_list)

double_columns = ['가격','종류','판매일자','반품여부']
pd.DataFrame(data=double_list, columns=double_columns)
```

### 내용 확인

```python
# @title
df_saledays = pd.DataFrame(data=double_list, columns=double_columns)
type(df_saledays), df_saledays.shape
```

```python
# @title
df_saledays.index, df_saledays.columns, df_saledays.values #dataframe 분해
```

```python
# @title
type(df_saledays.columns)  #index 기 때문에 바로 쓸 수 없음. 때문에 list로 cast 해야함
```

## 전처리

### 추출

```python
df_saledays
```

#### 컬럼 단위 추출

```python
df_saledays.columns
```

#### row(index) 단위 추출 : iloc vs loc

- iloc : 위치값으로 접근(index의 번호)
- loc : 명칭으로 접근(index의 이름)





```python
df_saledays
```

```python
df_saledays.index = ['a', 'b', 'c', 'd']
```

```python
df_saledays.iloc[2] , df_saledays.loc['c']
```

#### index와 columns 같이 추출

```python
df_saledays.loc['c', '판매일자'] # indexes, columns
```

```python
df_saledays.loc[:, '판매일자'] # 전체를 가지고 오기 위해
```

```python
df_saledays.loc[:, '판매일자':] # 전체를 가지고 오기 위해
```
