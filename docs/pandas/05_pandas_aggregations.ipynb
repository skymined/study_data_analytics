```python
import pandas as pd
```

```python
double_list = [
    [1000, '과자', '2019-12-31', '반품'],
    [2000, '음료', '2020-03-02', '정상'],
    [3000, '아이스크림', '2020-02-03', '정상'],
    [1000, '과자', '2019-12-31', '반품'],
    [4000, '초콜릿', '2020-03-15', '정상'],
    [5000, '사탕', '2020-04-01', '정상'],
    [6000, '빵', '2020-05-20', '반품'],
    [7000, '우유', '2020-06-30', '정상'],
    [8000, '치즈', '2020-07-07', '반품']
]
```

```python
double_columns = ['가격','종류','판매일자','반품여부']
pd.DataFrame(data=double_list, columns=double_columns)
```

```python
df_sale = pd.DataFrame(data=double_list, columns=double_columns)
df_sale
```

### pd.pivot_table()
- data : DataFrame
- index : group by (unique())
  - ex. 반품여부의 경우에는 반품과 정상 밖에 없으므로 index가 2개가 나옴
- value : aggreation 대상 컬럼명
- aggfunction : 집계(aggregation)

```python
# 각 group by의 row 개수 -> 빈도수. 반품을 얼마나 했는가!
pd.pivot_table(data=df_sale[['가격', '종류', '반품여부']]
               , index=['반품여부'], values=['가격'], aggfunc = 'count')
```

```python
# 가격이 합산 된 것을 보여줌
pd.pivot_table(data=df_sale[['가격', '종류', '반품여부']]
               , index=['반품여부'], values=['가격'], aggfunc = 'sum')
```

### pivot_table with numpy

```python
import numpy as np
pd.pivot_table(data=df_sale[['가격', '종류', '반품여부']]
               , index=['반품여부'], values=['가격'], aggfunc = np.sum)
```

```python
df_sales_pivots = pd.pivot_table(data=df_sale[['가격', '종류', '반품여부']]
               , index=['반품여부'], values=['가격'], aggfunc = np.sum)
```
